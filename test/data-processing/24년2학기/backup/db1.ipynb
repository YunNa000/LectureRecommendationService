### 1. 학정번호, year, semester, 교수명 넣기

csv 자체에 겹치는 값들이 존재해요.

set 써서 중복 없애고 진행할게요


```python
import pandas as pd
import sqlite3


csv_file_path = '2402강의계획서_전체_v1.csv'
db_file_path = 'kwu-lecture-database-v10.db'


df = pd.read_csv(csv_file_path, encoding="EUC-kr")
df_unique = df.drop_duplicates(subset='lecNumber')


conn = sqlite3.connect(db_file_path)
cursor = conn.cursor()

year = 24
semester = '2학기'

for index, row in df_unique.iterrows():
    lec_class_name = row['lecClassName'] if pd.notnull(
        row['lecClassName']) else ""
    lec_professor = row['lecProfessor'] if pd.notnull(
        row['lecProfessor']) else ""
    lec_classification = row['lecClassification'] if pd.notnull(
        row['lecClassification']) else ""
    lec_credit = row['lecCredit'] if pd.notnull(row['lecCredit']) else None
    lec_week_hours = row['lecWeekHours'] if pd.notnull(
        row['lecWeekHours']) else None

    cursor.execute('''
    INSERT INTO LectureTable (lecNumber, year, semester, lecClassName, lecProfessor, lecClassification, lecCredit, lecWeekHours)
    VALUES (?, ?, ?, ?, ?, ?, ?, ?)
    ''', (row['lecNumber'], year, semester, lec_class_name, lec_professor, lec_classification, lec_credit, lec_week_hours))

conn.commit()
conn.close()

print("done")
```

```python
import pandas as pd
import re


df = pd.read_csv('kwu-24summer-timetable_v4.csv', encoding="EUC-kr")


def transform_lecTime(lec_time):
    if isinstance(lec_time, str):

        matches = re.findall(r'\((\d+):(\d+)(?:,(\d+))?\)', lec_time)
        new_times = []

        for match in matches:
            first_part = match[0]
            second_part = match[1]

            new_times.append(f'({first_part}:{second_part})')
            if match[2]:
                new_times.append(f'({first_part}:{match[2]})')

        print(new_times)

        return ','.join(new_times)
    return ''


df['lecTime'] = df['lecTime'].apply(transform_lecTime)


df.to_csv('./kwu-24summer-timetable_v4.csv', index=False)
```

```python
import pandas as pd
import random
import string


csv_file_path = 'kwu-24summer-timetable_v4.csv'
data = pd.read_csv(csv_file_path, encoding="euc-kr")


def fill_lecNumber(row):
    if pd.isnull(row['lecNumber']) or row['lecNumber'] == '':
        random_number1 = random.randint(0, 9)
        random_alpha_numeric = ''.join(random.choices(
            string.ascii_letters + string.digits, k=4))
        random_number2 = random.randint(0, 9)
        return f"xxxx-{random_number1}-{random_alpha_numeric}-0{random_number2}"
    return row['lecNumber']


data['lecNumber'] = data.apply(fill_lecNumber, axis=1)


output_file_path = 'kwu-24summer-timetable_v4.csv'
data.to_csv(output_file_path, index=False, encoding="euc-kr")
```

```python
import sqlite3
import pandas as pd


conn = sqlite3.connect('kwu-lecture-recommendation-v2.db')
cursor = conn.cursor()


csv_file = 'kwu-24summer-timetable_v4.csv'
df = pd.read_csv(csv_file, encoding="euc-kr")


def get_lecture_id(year, semester, lec_number):
    cursor.execute("SELECT lectureID FROM LectureList WHERE year=? AND semester=? AND lecNumber=?",
                   (year, semester, lec_number))
    result = cursor.fetchone()
    return result[0] if result else None


for index, row in df.iterrows():
    year = row['yaer']
    semester = row['semester']
    lec_number = row['lecNumber']

    lecture_id = get_lecture_id(year, semester, lec_number)

    if lecture_id:
        can_take_bunban = row['canTakeBunBan'] if 'canTakeBunBan' in row else None
        major_recog_bunban = row['majorRecogBunBan'] if 'majorRecogBunBan' in row else None
        can_take_foreign_people = row['canTakeForeignPeople'] if 'canTakeForeignPeople' in row else None

        cursor.execute("""
            INSERT INTO LectureConditions (lectureID, canTakeBunBan, majorRecogBunBan, canTakeForeignPeople)
            VALUES (?, ?, ?, ?)
        """, (lecture_id, can_take_bunban, major_recog_bunban, can_take_foreign_people))


conn.commit()
conn.close()
```

```python
import sqlite3
import pandas as pd


conn = sqlite3.connect('kwu-lecture-recommendation-v2.db')
cursor = conn.cursor()


csv_file = 'kwu-24summer-timetable_v4.csv'
df = pd.read_csv(csv_file)


def get_lecture_id(year, semester, lec_number):
    cursor.execute("SELECT lectureID FROM LectureList WHERE year=? AND semester=? AND lecNumber=?",
                   (year, semester, lec_number))
    result = cursor.fetchone()
    return result[0] if result else None


for index, row in df.iterrows():
    year = row['yaer']
    semester = row['semester']
    lec_number = row['lecNumber']

    lecture_id = get_lecture_id(year, semester, lec_number)

    if lecture_id:
        taken_people_1_years_ago = row.get('takenPeople1yearsAgo', None)
        taken_people_2_years_ago = row.get('takenPeople2yearsAgo', None)
        taken_people_3_years_ago = row.get('takenPeople3yearsAgo', None)
        foreign_language = row.get('foreignLanguage', None)
        percentage_of_online = row.get('percentageOfOnline', None)
        is_pnp = row.get('isPNP', None)
        is_engineering = row.get('isEngineering', None)
        is_tbl = row.get('isTBL', None)
        is_pbl = row.get('isPBL', None)
        is_seminar = row.get('isSeminar', None)
        is_small = row.get('isSmall', None)
        is_convergence = row.get('isConvergence', None)
        is_team_teaching = row.get('isTeamTeaching', None)
        is_focus = row.get('isFocus', None)
        is_experiment_design = row.get('isExperimentDesign', None)
        is_e_learning = row.get('isELearning', None)
        is_art = row.get('isArt', None)
        represent_competency = row.get('representCompetency', None)
        learning_goal_n_method = row.get('learningGoalNmethod', None)
        overview = row.get('Overview', None)
        evaluation_ratio = row.get('evaluationRatio', None)
        main_book = row.get('mainBook', None)
        schedule_n_content = row.get('scheduleNcontent', None)

        cursor.execute("""
            INSERT INTO LectureDetailData (lectureID, takenPeople1yearsAgo, takenPeople2yearsAgo, takenPeople3yearsAgo, 
            ForeignLanguage, percentageOfOnline, isPNP, isEngineering, isTBL, isPBL, isSeminar, isSmall, 
            isConvergence, isTeamTeaching, isFocus, isExperimentDesign, isELearning, isArt, 
            representCompetency, learningGoalNmethod, Overview, evaluationRatio, mainBook, scheduleNcontent)
            VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
        """, (lecture_id, taken_people_1_years_ago, taken_people_2_years_ago, taken_people_3_years_ago,
              foreign_language, percentage_of_online, is_pnp, is_engineering, is_tbl, is_pbl, is_seminar,
              is_small, is_convergence, is_team_teaching, is_focus, is_experiment_design, is_e_learning,
              is_art, represent_competency, learning_goal_n_method, overview, evaluation_ratio, main_book,
              schedule_n_content))


conn.commit()


conn.close()
```

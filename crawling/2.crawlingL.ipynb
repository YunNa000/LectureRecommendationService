```python
import re

# 주어진 텍스트
text =page_content

# 정규 표현식을 사용하여 각 항목 추출하는 함수
def extract_info(pattern, text):
    match = re.search(pattern, text)
    return match.group(1).strip() if match else ""

# 항목별 정규 표현식 패턴
patterns = {
    "교과목명": r'교과목명\s(.*?)\s년도학기',
    "년도학기": r'년도학기\s(.*?)\s학정번호',
    "이수구분": r'이수구분\s(.*?)\s강의소개',
    "학정번호": r'학정번호\s(.*?)\s이수구분',
    "강의소개": r'강의소개\s(.*?)\s동영상',
    "학점/시간": r'학점/시간\s(.*?)\s강의시간',
    "강의시간 및 강의실": r'\(강의실\)\s(.*?)\s수강인원',
    "연락처": r'연락처\s(.*?)\s이동전화',
    "이동전화": r'이동전화\s(.*?)\s이메일',
    "이메일": r'이메일\s(.*?)\n'
}



# 각 항목 추출
course_name = extract_info(patterns["교과목명"], text)
year_semester = extract_info(patterns["년도학기"], text)
course_classification = extract_info(patterns["이수구분"], text)
course_code = extract_info(patterns["학정번호"], text)
course_description = extract_info(patterns["강의소개"], text)
credit_time = extract_info(patterns["학점/시간"], text)
class_time_location = extract_info(patterns["강의시간 및 강의실"], text)
contact_number = extract_info(patterns["연락처"], text)
mobile_number = extract_info(patterns["이동전화"], text)
email = extract_info(patterns["이메일"], text)

# 결과 출력
print("교과목명:", course_name)
print("년도학기:", year_semester)
print("이수구분:", course_classification)
print("학정번호:", course_code)
print("강의소개:", course_description)
print("학점/시간:", credit_time)
print("강의시간 및 강의실:", class_time_location)
print("연락처:", contact_number)
print("이동전화:", mobile_number)
print("이메일:", email)
```

```python


# 정규 표현식을 사용하여 각 항목 추출하는 함수
def extract_info(pattern, text):
    match = re.search(pattern, text, re.DOTALL)
    return match.group(1).strip() if match else ""

# 항목별 정규 표현식 패턴
patterns = {
    "담당조교": r'담당조교\s(.*?)\s',
    "장애학생 지원": r'장애학생 지원\s(.*?)\s교과목',
    "교과목 개요": r'개요\s(.*?)\s교과목\s대표역량',
    "대표역량": r'대표역량\s(.*?)\s학습목표 및',
    "학습목표 및 학습방법": r'학습목표 및\s학습방법\s(.*?)\s교과목',
    "학습성과": r'학습성과\s(.*?)\s강의유형',
}

# 각 항목 추출
assistant_name = extract_info(patterns["담당조교"], text)
support_for_disabled_students = extract_info(patterns["장애학생 지원"], text)
course_overview = extract_info(patterns["교과목 개요"], text)
core_competency = extract_info(patterns["대표역량"], text)
learning_objectives_methods = extract_info(patterns["학습목표 및 학습방법"], text)
learning_outcomes = extract_info(patterns["학습성과"], text)

# 결과 출력
print("담당조교:", assistant_name)
print("장애학생 지원:", support_for_disabled_students)
print("교과목 개요:", course_overview)
print("대표역량:", core_competency)
print("학습목표 및 학습방법:", learning_objectives_methods)
print("학습성과:", learning_outcomes)
```

```python


# 정규 표현식을 사용하여 각 항목 추출하는 함수
def extract_info(pattern, text):
    match = re.search(pattern, text)
    return match.group(1).strip() if match else ""

# 영어강의 비율 추출
english_lecture_rate = extract_info(r'영어강의 비율\s(\d+ %)', text)

# 평가방법 비율 추출
evaluation_methods = extract_info(r'평가방법 비율\n출석 중간고사 기말고사 과제보고서 수업태도 Quiz 기타\n(.*?)\n기타평가', text)
evaluation_values = evaluation_methods.split()

# 각 평가 항목에 대한 값을 딕셔너리에 저장
evaluation_dict = {
    "출석": evaluation_values[0],
    "중간고사": evaluation_values[1],
    "기말고사": evaluation_values[2],
    "과제보고서": evaluation_values[3],
    "수업태도": evaluation_values[4],
    "Quiz": evaluation_values[5],
    "기타": evaluation_values[6]
}

# 결과 출력
print("영어강의 비율:", english_lecture_rate)
print("평가방법 비율:")
for key, value in evaluation_dict.items():
    print(f"{key}: {value}")
```

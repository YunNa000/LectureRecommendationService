```python
import sqlite3
from sentence_transformers import SentenceTransformer, util


model = SentenceTransformer('intfloat/multilingual-e5-large-instruct')


conn = sqlite3.connect('./kwu-lecture-recommendation-v7.db')
cursor = conn.cursor()


cursor.execute(
    """
SELECT lectureID, ReviewSummary
FROM LectureDetailData s
WHERE ReviewSummary IS NOT NULL AND ReviewSummary <> ''
""")
rows = cursor.fetchall()


for lectureID, ReviewSummary in rows:
    EverytimeEmbedding = model.encode(ReviewSummary)
    cursor.execute(
        "UPDATE LectureDetailData SET EverytimeEmbedding = ? WHERE lectureID = ?", (EverytimeEmbedding, lectureID))

    print(lectureID, EverytimeEmbedding)

conn.commit()
conn.close()
```

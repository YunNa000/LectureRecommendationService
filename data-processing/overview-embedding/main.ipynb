```python
import sqlite3
from sentence_transformers import SentenceTransformer, util


model = SentenceTransformer('intfloat/multilingual-e5-large-instruct')


conn = sqlite3.connect('./kwu-lecture-recommendation-v6.db')
cursor = conn.cursor()


cursor.execute(
    """
SELECT lectureID, Overview
FROM LectureDetailData s
WHERE Overview IS NOT NULL AND Overview <> ''
""")
rows = cursor.fetchall()


for lectureID, Overview in rows:
    overview_embeddings = model.encode(Overview)
    cursor.execute(
        "UPDATE LectureDetailData SET OverviewEmbedding = ? WHERE lectureID = ?", (overview_embeddings, lectureID))

    print(lectureID, overview_embeddings)

conn.commit()
conn.close()
```

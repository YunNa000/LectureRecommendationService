````python
import os
from pdf2image import convert_from_path
from PIL import Image


pdf_dir = './pdfs'
img_dir = './imgs'


os.makedirs(img_dir, exist_ok=True)


for filename in os.listdir(pdf_dir):
    if filename.endswith('.pdf'):
        pdf_path = os.path.join(pdf_dir, filename)

        images = convert_from_path(pdf_path)

        total_width = max(image.width for image in images)
        total_height = sum(image.height for image in images)

        combined_image = Image.new('RGB', (total_width, total_height))

        y_offset = 0
        for image in images:
            combined_image.paste(image, (0, y_offset))
            y_offset += image.height

        img_path = os.path.join(
            img_dir, f"{os.path.splitext(filename)[0]}.jpg")
        combined_image.save(img_path)

print("done")
````

````python
import os
from PIL import Image
from reportlab.pdfgen import canvas


def image_to_pdf(image_path, pdf_path):
    img = Image.open(image_path)
    width, height = img.size

    pdf_width = width
    pdf_height = height

    c = canvas.Canvas(pdf_path, pagesize=(pdf_width, pdf_height))
    c.drawImage(image_path, 0, 0, width=pdf_width, height=pdf_height)
    c.showPage()
    c.save()


def convert_all_images_to_pdf(input_directory, output_directory):
    # 출력 디렉토리가 없으면 생성
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    # 디렉토리 내의 모든 파일에 대해 반복
    for filename in os.listdir(input_directory):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):  # 이미지 파일 확장자
            image_path = os.path.join(input_directory, filename)
            pdf_filename = os.path.splitext(
                filename)[0] + '.pdf'  # PDF 파일 이름 생성
            pdf_path = os.path.join(output_directory, pdf_filename)

            # 이미지 파일을 PDF로 변환
            image_to_pdf(image_path, pdf_path)
            print(f'Converted {filename} to {pdf_filename}')


# 사용 예시
input_directory = './imgs'         # 입력 이미지가 있는 디렉토리
output_directory = './final_pdfs'  # 출력 PDF가 저장될 디렉토리
convert_all_images_to_pdf(input_directory, output_directory)
````

````python
import os
import re
from openai import OpenAI
from openai.types.beta.threads.message_create_params import Attachment, AttachmentToolFileSearch
import PRIVATE

client = OpenAI(
    api_key=PRIVATE.api_key
)


def create_file(file_path):
    with open(file_path, 'rb') as f:
        return client.files.create(
            file=f,
            purpose='assistants'
        )


def get_assistant():
    for assistant in client.beta.assistants.list():
        if assistant.name == 'My Assistant Name':
            return assistant

    return client.beta.assistants.create(
        model='gpt-4o-mini',
        description="You must be convert tables and text that appear in a pdf to markdown. When converting to markdown, you never use two spaces in a row or -- twice in a row.",
        instructions="너는 pdf의 전체 내용을 markdown으로 변환하고, 오직 markdown으로만 답을 하는 제일 도움되는 assistant야.",
        tools=[{"type": "file_search"}],
        name='My Assistant Name',
    )


def process_pdf(file_path, assistant_id):
    file = create_file(file_path)
    thread = client.beta.threads.create()

    prompt = "해당 pdf의 표를 markdown 표와 텍스트들을 markdown으로 변환해."
    client.beta.threads.messages.create(
        thread_id=thread.id,
        role='user',
        content=prompt,
        attachments=[Attachment(file_id=file.id, tools=[
                                AttachmentToolFileSearch(type='file_search')])]
    )

    run = client.beta.threads.runs.create_and_poll(
        thread_id=thread.id,
        assistant_id=assistant_id,
        timeout=300,
    )

    if run.status != "completed":
        raise Exception('Run failed:', run.status)

    messages_cursor = client.beta.threads.messages.list(thread_id=thread.id)
    messages = [message for message in messages_cursor]

    message = messages[0]
    assert message.content[0].type == "text"

    res_txt = message.content[0].text.value

    if res_txt.startswith('```markdown'):
        res_txt = res_txt[11:]
    if res_txt.endswith('```'):
        res_txt = res_txt[:-3]

    client.files.delete(file.id)
    return res_txt


def normalize_spaces(text):
    normalized_text = re.sub(r'\s{2,}', ' ', text)
    normalized_text2 = re.sub(r'\s{2,}', '-', normalized_text)
    return normalized_text2


def save_as_md(file_name, content):
    with open(file_name, 'w', encoding='utf-8') as file:
        file.write(content)


def main():
    pdf_dir = './final_pdfs/'
    md_dir = './md/'
    os.makedirs(md_dir, exist_ok=True)

    assistant = get_assistant()

    for pdf_file in os.listdir(pdf_dir):
        if pdf_file.endswith('.pdf'):
            pdf_path = os.path.join(pdf_dir, pdf_file)
            md_file_name = os.path.splitext(pdf_file)[0] + '.md'
            md_path = os.path.join(md_dir, md_file_name)

            try:
                res_txt = process_pdf(pdf_path, assistant.id)
                normalized_res_txt = normalize_spaces(res_txt)
                save_as_md(md_path, normalized_res_txt)
                print(f'{pdf_file} done')
            except Exception as e:
                print(f'err {pdf_file}: {e}')


if __name__ == "__main__":
    main()
````

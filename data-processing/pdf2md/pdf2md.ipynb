````python
import PyPDF2


def extract_pages(input_pdf_path, output_pdf_path, start_page, end_page):
    with open(input_pdf_path, 'rb') as input_file:
        reader = PyPDF2.PdfReader(input_file)
        writer = PyPDF2.PdfWriter()
        output_pdf_path = "./pdfs/"+output_pdf_path

        for page_num in range(start_page-1, end_page):
            page = reader.pages[page_num]
            writer.add_page(page)

        with open(output_pdf_path, 'wb') as output_file:
            writer.write(output_file)

    print(f'{output_pdf_path} done')


extract_pages('2-1. 수강신청_자료집_전체(2024-2).pdf', '수강신청.pdf', 8, 18)

extract_pages('2-1. 수강신청_자료집_전체(2024-2).pdf',
              '졸업이수학점 안내 (19년 이후 신입학자).pdf', 26, 31) # (일반_편입생)졸업이수학점_안내

extract_pages('2-1. 수강신청_자료집_전체(2024-2).pdf',
              '졸업이수학점 안내 (공학인증학과, 19년 이후 신입학자).pdf', 44, 46) # (공학인증학과)졸업이수학점_안내

extract_pages('2-1. 수강신청_자료집_전체(2024-2).pdf',
              '졸업이수학점 안내 (다전공, 19년 이후 신입학자).pdf', 37, 39) # (다전공)졸업이수학점_안내

extract_pages('2-1. 수강신청_자료집_전체(2024-2).pdf', '교양 필수 교과목 수강 안내.pdf', 32, 33)

extract_pages('2-1. 수강신청_자료집_전체(2024-2).pdf',
              '편입생이 이수해야 할 교과목 및 학점.pdf', 34, 34)

extract_pages('2-1. 수강신청_자료집_전체(2024-2).pdf', '다전공 안내.pdf', 34, 39)

extract_pages('2-1. 수강신청_자료집_전체(2024-2).pdf', '연계전공 안내.pdf', 40, 41)

extract_pages('2-1. 수강신청_자료집_전체(2024-2).pdf', '공학교육인증제.pdf', 42, 46)

extract_pages('2-1. 수강신청_자료집_전체(2024-2).pdf', '학-석사 연계과정.pdf', 79, 80)

extract_pages('2-1. 수강신청_자료집_전체(2024-2).pdf', '서비스러닝 교과목 안내.pdf', 81, 82)

extract_pages('2-1. 수강신청_자료집_전체(2024-2).pdf',
              '다학년 다학기 프로젝트 (KW-VIP) 안내.pdf', 83, 86)

extract_pages('2-1. 수강신청_자료집_전체(2024-2).pdf', '참빛설계학기 안내.pdf', 87, 88)

extract_pages('2-1. 수강신청_자료집_전체(2024-2).pdf',
              'K-MOOC 교내 온라인강좌 수강 안내.pdf', 89, 89)

extract_pages('2-1. 수강신청_자료집_전체(2024-2).pdf',
              '매치업 집중이수제 교내 강좌 수강 안내.pdf', 90, 90)

extract_pages('2-1. 수강신청_자료집_전체(2024-2).pdf', '현장실습학기제 안내.pdf', 91, 91)

extract_pages('2-1. 수강신청_자료집_전체(2024-2).pdf',
              '특별교육과정 (CDP, R.O.T.C., 글로벌인재트랙인증제).pdf', 92, 95)
````

````python
import os
from pdf2image import convert_from_path
from PIL import Image


pdf_dir = './pdfs'
img_dir = './imgs'


os.makedirs(img_dir, exist_ok=True)


for filename in os.listdir(pdf_dir):
    if filename.endswith('.pdf'):
        pdf_path = os.path.join(pdf_dir, filename)

        images = convert_from_path(pdf_path)

        total_width = max(image.width for image in images)
        total_height = sum(image.height for image in images)

        combined_image = Image.new('RGB', (total_width, total_height))

        y_offset = 0
        for image in images:
            combined_image.paste(image, (0, y_offset))
            y_offset += image.height

        img_path = os.path.join(
            img_dir, f"{os.path.splitext(filename)[0]}.jpg")
        combined_image.save(img_path)

print("done")
````

````python
import os
from PIL import Image
from reportlab.pdfgen import canvas


def image_to_pdf(image_path, pdf_path):
    img = Image.open(image_path)
    width, height = img.size

    pdf_width = width
    pdf_height = height

    c = canvas.Canvas(pdf_path, pagesize=(pdf_width, pdf_height))
    c.drawImage(image_path, 0, 0, width=pdf_width, height=pdf_height)
    c.showPage()
    c.save()


def convert_all_images_to_pdf(input_directory, output_directory):
    # 출력 디렉토리가 없으면 생성
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    # 디렉토리 내의 모든 파일에 대해 반복
    for filename in os.listdir(input_directory):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):  # 이미지 파일 확장자
            image_path = os.path.join(input_directory, filename)
            pdf_filename = os.path.splitext(
                filename)[0] + '.pdf'  # PDF 파일 이름 생성
            pdf_path = os.path.join(output_directory, pdf_filename)

            # 이미지 파일을 PDF로 변환
            image_to_pdf(image_path, pdf_path)
            print(f'Converted {filename} to {pdf_filename}')


# 사용 예시
input_directory = './imgs'         # 입력 이미지가 있는 디렉토리
output_directory = './final_pdfs'  # 출력 PDF가 저장될 디렉토리
convert_all_images_to_pdf(input_directory, output_directory)
````

````python
import os
import re
from google.cloud import documentai_v1 as documentai
from google.cloud import language_v1
# from google.cloud import storage
from google.oauth2 import service_account

os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = r"C:\Users\user\seventh-palace-431311-s0-790c4d383ee1.json"

key_path=r"C:/Users/user\seventh-palace-431311-s0-790c4d383ee1.json"

# 자격 증명 설정
credentials = service_account.Credentials.from_service_account_file(key_path)

# Google Cloud 프로젝트 설정
project_id = "seventh-palace-431311-s0"
location = "us"  # 예: "us" 또는 "eu"
processor_id = "3c489a1e3d53f26b"  # Document AI 프로세서 ID

def process_document(file_path):
    client = documentai.DocumentProcessorServiceClient()
    name = f"projects/{project_id}/locations/{location}/processors/{processor_id}"

    with open(file_path, "rb") as image:
        image_content = image.read()

    document = {"content": image_content, "mime_type": "application/pdf"}
    request = documentai.ProcessRequest(name=name, raw_document=document)
    result = client.process_document(request=request)
    return result.document

def analyze_text(text):
    client = language_v1.LanguageServiceClient()
    document = languagev1.Document(content=text, type=language_v1.Document.Type.PLAIN_TEXT)
    response = client.analyze_entities(document=document)
    return response.entities

def convert_to_markdown(document):
    md_content = ""
    for page in document.pages:
        for table in page.tables:
            md_table = "| " + " | ".join(["Header"] * len(table.header_rows[0].cells)) + " |\n"
            md_table += "| " + " | ".join(["---"] * len(table.header_rows[0].cells)) + " |\n"
            for row in table.body_rows:
                md_table += "| " + " | ".join([cell.layout.text_anchor.content for cell in row.cells]) + " |\n"
            md_content += md_table + "\n\n"

        for paragraph in page.paragraphs:
            md_content += paragraph.layout.text_anchor.content + "\n\n"

    return md_content

def normalize_spaces(text):
    normalized_text = re.sub(r'\s{2,}', ' ', text)
    normalized_text = re.sub(r'(?<!\n)\n(?!\n)', ' ', normalized_text)  # 단일 줄바꿈을 공백으로
    return normalized_text

def save_as_md(file_name, content):
    with open(file_name, 'w', encoding='utf-8') as file:
        file.write(content)

def process_pdf(file_path, output_path):
    try:
        document = process_document(file_path)
        md_content = convert_to_markdown(document)

        # NLP 분석 (선택적)
        # entities = analyze_text(md_content)
        # 여기서 entities를 사용하여 추가적인 처리를 할 수 있습니다.

        normalized_content = normalize_spaces(md_content)
        save_as_md(output_path, normalized_content)
        print(f"Converted {file_path} to {output_path}")
    except Exception as e:
        print(f"Error processing {file_path}: {str(e)}")

def main():
    pdf_dir = './final_pdfs/'
    md_dir = './md/'
    os.makedirs(md_dir, exist_ok=True)

    for pdf_file in os.listdir(pdf_dir):
        if pdf_file.endswith('.pdf'):
            pdf_path = os.path.join(pdf_dir, pdf_file)
            md_file_name = os.path.splitext(pdf_file)[0] + '.md'
            md_path = os.path.join(md_dir, md_file_name)

            process_pdf(pdf_path, md_path)

if __name__ == "__main__":
    main()
````

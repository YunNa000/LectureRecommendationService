```python
import sqlite3

DATABASE = './kwu-lecture-database-v8.db'

def db_connect():
    conn = sqlite3.connect(DATABASE)
    conn.row_factory = sqlite3.Row
    return conn

def get_lecture_table_schema():
    conn = db_connect()
    cursor = conn.cursor()
    
    # LectureTable의 스키마 정보를 가져오는 쿼리
    cursor.execute("PRAGMA table_info(LectureTable)")
    
    # 결과 출력
    print("LectureTable 스키마:")
    print("--------------------")
    for column in cursor.fetchall():
        print(f"칼럼명: {column['name']}, 데이터 타입: {column['type']}")
    
    conn.close()

if __name__ == "__main__":
    get_lecture_table_schema()
```

### 2차 전처리
* 1. 중복값처리 ex. 매치업 (교양파일과 전공파일에 모두 존재)

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\user\Desktop\df_result.csv", encoding="cp949")
```

```python
# 중복값과 그 개수 찾기
duplicate_counts = df['학정번호'].value_counts()
duplicates = duplicate_counts[duplicate_counts > 1]
print("중복값과 개수:")
print(duplicates)

# 중복값 중 하나만 남기고 삭제하기
df_unique = df.drop_duplicates(subset='학정번호', keep='first')

print("\n중복값 중 하나만 남기고 삭제한 데이터프레임:")
print(df_unique)

# 데이터프레임을 CSV 파일로 저장하기
csv_file_path = 'C:/Users/user/Desktop/unique_학정번호1.csv'
df_unique.to_csv(csv_file_path, index=False, encoding='cp949')

print("\n파일로 저장되었습니다.")
```

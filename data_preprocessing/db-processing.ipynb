```python
conn = sqlite3.connect('./kwu-lecture-database.db')
cursor = conn.cursor()

df = pd.read_csv("./kwu-lecture-timetable.csv")

lecNumLists = df['학정번호'].tolist()

query = '''
SELECT lecNumber
FROM LectureTable
WHERE lecNumber IN ({})
AND lecUrl IS NULL
'''.format(','.join(['?']*len(lecNumLists)))

cursor.execute(query, lecNumLists)
result = cursor.fetchall()

lecNumWithoutUrl = [row[0] for row in result]

conn.close()

print(lecNumWithoutUrl)
print(len(lecNumWithoutUrl))
```

```python
import pandas as pd


df = pd.read_csv("./kwu-lecture-timetable.csv")


dups = df['학정번호'].duplicated(keep=False)

dup = df[dups]

grouped = dup.groupby('학정번호')

for 학정번호, group in grouped:
    df.drop(group.index[-1], inplace=True)

df.to_csv('final.csv', index=False, encoding="cp949")
```

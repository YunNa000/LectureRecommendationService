```python
import pandas as pd
import sqlite3

df = pd.read_csv("./kwu-lecture-timetable.csv")

conn = sqlite3.connect('./kwu-lecture-database.db')
cursor = conn.cursor()

def process_years(years):
    year_dict = {
        'lecTakeOnly1Year': 0,
        'lecTakeOnly2Year': 0,
        'lecTakeOnly3Year': 0,
        'lecTakeOnly4Year': 0,
        'lecTakeOnly5Year': 0,
    }
    if pd.isna(years):
        return year_dict
    for year in years.split(','):
        year = year.strip()
        if year == '1':
            year_dict['lecTakeOnly1Year'] = 1
        elif year == '2':
            year_dict['lecTakeOnly2Year'] = 1
        elif year == '3':
            year_dict['lecTakeOnly3Year'] = 1
        elif year == '4':
            year_dict['lecTakeOnly4Year'] = 1
        elif year == '5':
            year_dict['lecTakeOnly5Year'] = 1
    return year_dict

for index, row in df.iterrows():
    years = row['학년']
    year_data = process_years(years)
    print(f"{years} | {row['학정번호']} | {year_data['lecTakeOnly1Year']} | {year_data['lecTakeOnly2Year']} | {year_data['lecTakeOnly3Year']} | {year_data['lecTakeOnly4Year']} | {year_data['lecTakeOnly5Year']}")
    
    cursor.execute('''
    UPDATE LectureTable
    SET lecTakeOnly1Year = ?, lecTakeOnly2Year = ?, lecTakeOnly3Year = ?, lecTakeOnly4Year = ?, lecTakeOnly5Year = ?
    WHERE lecNumber = ?
    ''', (year_data['lecTakeOnly1Year'], year_data['lecTakeOnly2Year'], year_data['lecTakeOnly3Year'], year_data['lecTakeOnly4Year'], year_data['lecTakeOnly5Year'], row['학정번호']))

conn.commit()
conn.close()
```

```python
import pandas as pd
import sqlite3

df = pd.read_csv("./kwu-lecture-timetable.csv")

conn = sqlite3.connect('./kwu-lecture-database.db')
cursor = conn.cursor()

def process_yi_su(yi_su):
    year_dict = {'lecClassification': None}
    if pd.isna(yi_su):
        return year_dict
    for i in yi_su.split(','):
        i = i.strip()
        if i == '교필':
            year_dict['lecClassification'] = "교필"
        elif i == '교선':
            year_dict['lecClassification'] = "전필"
        elif i == '교선':
            year_dict['lecClassification'] = "기선"
        elif i == '교선':
            year_dict['lecClassification'] = "무관"
        elif i == '교선':
            year_dict['lecClassification'] = "일선"
        elif i == '교선':
            year_dict['lecClassification'] = "교선"
        elif i == '교선':
            year_dict['lecClassification'] = "전선"

    return year_dict

for index, row in df.iterrows():
    yi_su = row['이수']
    yi_su_data = process_yi_su(yi_su)
    print(f"{yi_su} | {row['학정번호']} | {yi_su_data['lecClassification']}")
    
    cursor.execute('''
    UPDATE LectureTable
    SET lecClassification = ?
    WHERE lecNumber = ?
    ''', (yi_su_data['lecClassification'], row['학정번호']))

conn.commit()
conn.close()
```

```python
import pandas as pd
import sqlite3

df = pd.read_csv("./kwu-lecture-timetable.csv")

conn = sqlite3.connect('./kwu-lecture-database.db')
cursor = conn.cursor()

cursor.execute('''
ALTER TABLE LectureTable
ADD COLUMN lecCredit INTEGER;
''')

def process_yi_su(yi_su):
    year_dict = {'lecCredit': None}
    if pd.isna(yi_su):
        return year_dict
    yi_su_str = str(yi_su).strip()
    if yi_su_str.isdigit():
        year_dict['lecCredit'] = int(yi_su_str)
    return year_dict

for index, row in df.iterrows():
    yi_su = row['학점']
    yi_su_data = process_yi_su(yi_su)
    print(f"{yi_su} | {row['학정번호']} | {yi_su_data['lecCredit']}")
    
    cursor.execute('''
    UPDATE LectureTable
    SET lecCredit = ?
    WHERE lecNumber = ?
    ''', (yi_su_data['lecCredit'], row['학정번호']))

conn.commit()
conn.close()
```

짜피 우리가 생각했던 기존 로직에 의해서

후수과목과 동일한 이름을 가지지만, 들을 수 없는 과목들은 걸러지겠구나


```python
import csv
import sqlite3

conn = sqlite3.connect('./kwu-lecture-database.db')
cursor = conn.cursor()

with open('./kwu-required-lecture.csv', 'r') as csv_file:
    csv_reader = csv.reader(csv_file)
    next(csv_reader)  
    for row in csv_reader:
        bunban = row[0].split(',')
        prerequisite = row[1]
        requirement = row[2]

        for b in bunban:
            cursor.execute("SELECT lecClassName FROM LectureTable WHERE lecCanTakeBunban LIKE ?", (f'%{b}%',))
            lecture_classes = [row[0] for row in cursor.fetchall()]
            if requirement in lecture_classes:
                idx = lecture_classes.index(requirement)
                cursor.execute("UPDATE LectureTable SET lecRequirementClass = ? WHERE lecClassName = ?", (prerequisite, lecture_classes[idx]))

conn.commit()
conn.close()
```

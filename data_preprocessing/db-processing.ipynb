```python
import pandas as pd
import sqlite3

df = pd.read_csv("./kwu-lecture-timetable.csv")

conn = sqlite3.connect('./kwu-lecture-database.db')
cursor = conn.cursor()

def process_years(years):
    year_dict = {
        'lecTakeOnly1Year': 0,
        'lecTakeOnly2Year': 0,
        'lecTakeOnly3Year': 0,
        'lecTakeOnly4Year': 0,
        'lecTakeOnly5Year': 0,
    }
    if pd.isna(years):
        return year_dict
    for year in years.split(','):
        year = year.strip()
        if year == '1':
            year_dict['lecTakeOnly1Year'] = 1
        elif year == '2':
            year_dict['lecTakeOnly2Year'] = 1
        elif year == '3':
            year_dict['lecTakeOnly3Year'] = 1
        elif year == '4':
            year_dict['lecTakeOnly4Year'] = 1
        elif year == '5':
            year_dict['lecTakeOnly5Year'] = 1
    return year_dict

for index, row in df.iterrows():
    years = row['학년']
    year_data = process_years(years)
    print(f"{years} | {row['학정번호']} | {year_data['lecTakeOnly1Year']} | {year_data['lecTakeOnly2Year']} | {year_data['lecTakeOnly3Year']} | {year_data['lecTakeOnly4Year']} | {year_data['lecTakeOnly5Year']}")
    
    cursor.execute('''
    UPDATE LectureTable
    SET lecTakeOnly1Year = ?, lecTakeOnly2Year = ?, lecTakeOnly3Year = ?, lecTakeOnly4Year = ?, lecTakeOnly5Year = ?
    WHERE lecNumber = ?
    ''', (year_data['lecTakeOnly1Year'], year_data['lecTakeOnly2Year'], year_data['lecTakeOnly3Year'], year_data['lecTakeOnly4Year'], year_data['lecTakeOnly5Year'], row['학정번호']))

conn.commit()
conn.close()
```

```python
import pandas as pd
import sqlite3

df = pd.read_csv("./kwu-lecture-timetable.csv")

conn = sqlite3.connect('./kwu-lecture-database.db')
cursor = conn.cursor()

def process_yi_su(yi_su):
    year_dict = {'lecClassification': None}
    if pd.isna(yi_su):
        return year_dict
    for i in yi_su.split(','):
        i = i.strip()
        if i == '교필':
            year_dict['lecClassification'] = "교필"
        elif i == '교선':
            year_dict['lecClassification'] = "전필"
        elif i == '교선':
            year_dict['lecClassification'] = "기선"
        elif i == '교선':
            year_dict['lecClassification'] = "무관"
        elif i == '교선':
            year_dict['lecClassification'] = "일선"
        elif i == '교선':
            year_dict['lecClassification'] = "교선"
        elif i == '교선':
            year_dict['lecClassification'] = "전선"

    return year_dict

for index, row in df.iterrows():
    yi_su = row['이수']
    yi_su_data = process_yi_su(yi_su)
    print(f"{yi_su} | {row['학정번호']} | {yi_su_data['lecClassification']}")
    
    cursor.execute('''
    UPDATE LectureTable
    SET lecClassification = ?
    WHERE lecNumber = ?
    ''', (yi_su_data['lecClassification'], row['학정번호']))

conn.commit()
conn.close()
```

```python
import pandas as pd
import sqlite3

df = pd.read_csv("./kwu-lecture-timetable.csv")

conn = sqlite3.connect('./kwu-lecture-database.db')
cursor = conn.cursor()

cursor.execute('''
ALTER TABLE LectureTable
ADD COLUMN lecCredit INTEGER;
''')

def process_yi_su(yi_su):
    year_dict = {'lecCredit': None}
    if pd.isna(yi_su):
        return year_dict
    yi_su_str = str(yi_su).strip()
    if yi_su_str.isdigit():
        year_dict['lecCredit'] = int(yi_su_str)
    return year_dict

for index, row in df.iterrows():
    yi_su = row['학점']
    yi_su_data = process_yi_su(yi_su)
    print(f"{yi_su} | {row['학정번호']} | {yi_su_data['lecCredit']}")
    
    cursor.execute('''
    UPDATE LectureTable
    SET lecCredit = ?
    WHERE lecNumber = ?
    ''', (yi_su_data['lecCredit'], row['학정번호']))

conn.commit()
conn.close()
```

```python
import csv
import sqlite3

conn = sqlite3.connect('./kwu-lecture-database.db')
cursor = conn.cursor()

with open('./kwu-required-lecture.csv', 'r') as csv_file:
    csv_reader = csv.reader(csv_file)
    next(csv_reader)  
    for row in csv_reader:
        bunban = row[0].split(',')
        prerequisite = row[1]
        requirement = row[2]

        for b in bunban:
            cursor.execute("SELECT lecClassName FROM LectureTable WHERE lecCanTakeBunban LIKE ?", (f'%{b}%',))
            lecture_classes = [row[0] for row in cursor.fetchall()]
            if requirement in lecture_classes:
                idx = lecture_classes.index(requirement)
                cursor.execute("UPDATE LectureTable SET lecRequirementClass = ? WHERE lecClassName = ?", (prerequisite, lecture_classes[idx]))

conn.commit()
conn.close()
```

```python
import csv
import sqlite3

with open('lecture_plan_v1.csv', 'r') as csv_file:
    reader = csv.DictReader(csv_file)
    data = list(reader)

conn = sqlite3.connect('kwu-lecture-database-v2.db')
c = conn.cursor()

for row in data:
    lecture_number = row['학정번호']
        
    c.execute("""
        UPDATE LectureTable
        SET
            lecIsEng = ?,
            lecIsCn = ?,
            lecIsJp = ?,
            lecIsTBL = ?,
            lecIsPBL = ?,
            lecIsSeminar = ?,
            lecIsSmall = ?,
            lecIsConvergence = ?,
            lecIsTeamTeach = ?,
            lecIsWorkStudyCombine = ?,
            lecIseLearning = ?,
            lecIsExperimentDesign = ?,
            lecIsFocus = ?,
            lecIsDistance100 = ?,
            lecIsDistance50 = ?,
            lecIsFace = ?,
            lecIsOnline = ?,
            lecIsRecorded = ?,
            lecIsDoExperiment = ?,
            lecIsDiscussion = ?,
            lecIsArt = ?
        WHERE lecNumber = ?
    """, (
        row['영어강의'] == 'checked',
        row['중국어 강의'] == 'checked',
        row['일본어 강의'] == 'checked',
        row['TBL강의'] == 'checked',
        row['PBL강의'] == 'checked',
        row['세미나강의'] == 'checked',
        row['소규모강의'] == 'checked',
        row['융합강의'] == 'checked',
        row['팀티칭강의'] == 'checked',
        row['일학습병행강의'] == 'checked',
        row['서울권역e-러닝'] == 'checked',
        row['실험실습실기설계강의'] == 'checked',
        row['집중이수제강의'] == 'checked',
        row['원격수업 (100%)'] == 'checked',
        row['원격수업 (50% 이상)'] == 'checked',
        row['100%대면강의'] == 'checked',
        row['100%실시간 화상강의'] == 'checked' or row['대면+실시간 화상강의'] == 'checked' or row['대면+실시간 화상강의+사전녹화강의'] == 'checked',
        row['100%사전녹화강의'] == 'checked' or row['대면+사전녹화강의'] == 'checked' or row['대면+실시간 화상강의+사전녹화강의'] == 'checked',
        row['실험/실습 실기'] == 'checked',
        row['토론식 수업(세미나 포함)프로젝트'] == 'checked',
        row['실기(예체능)'] == 'checked',
        lecture_number
    ))

conn.commit()
conn.close()
```

```python
import sqlite3

db_file = './kwu-lecture-database-v3.db'

target_lec_names = [
    '광운인되기',
    '21세기기업의인재상',
    '사회봉사1',
    'TOEIC연습1',
    'OPIc연습',
    '심층탐구인간과세계의이해',
    '경력개발과취업전략',
    '전공미리보기(OT학기제)',
    '진로탐색및설계'
]

conn = sqlite3.connect(db_file)
c = conn.cursor()

update_query = """
UPDATE LectureTable
SET LecIsPNP = 1
WHERE LecClassName IN ({})
"""
placeholder = ', '.join(['?'] * len(target_lec_names))
c.execute(update_query.format(placeholder), target_lec_names)

conn.commit()

conn.close()
```

```python
import sqlite3

db_file = './kwu-lecture-database-v3.db'

result_set = set()

conn = sqlite3.connect(db_file)
c = conn.cursor()

query = "SELECT lecRequirementClass, lecClassName FROM LectureTable WHERE lecRequirementClass IS NOT NULL AND lecRequirementClass != ''"
c.execute(query)

for row in c.fetchall():
    result_set.add(row[0])
    result_set.add(row[1])

conn.close()

print(result_set)
```

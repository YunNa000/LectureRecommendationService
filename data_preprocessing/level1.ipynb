## Preprocessing_1차데이터전처리
* 1. 페이지 사이 빈칸 삭제
* 2. NaN값을 NULL로 처리

```python
import pandas as pd

df1 = pd.read_csv(r"C:\Users\user\Desktop\3-2. 교양,매치업,군사학.csv", encoding="cp949")
df2 = pd.read_csv(r"C:\Users\user\Desktop\3-3. 연계,창업,융합,kw-tip.csv", encoding="cp949")
df3 = pd.read_csv(r"C:\Users\user\Desktop\3-4. 전공(전체).csv", encoding="cp949")
```

### 1. 페이지 사이의 빈칸 없애기

```python
df1 = df1.dropna(how='all') # 교양 page.20
df2 = df2.dropna(how='all') # 연계창업융합 page.2
df3 = df3.dropna(how='all') # 전공 page.40
```

```python
num_rows1, num_columns1 = df1.shape
num_rows2, num_columns2 = df2.shape
num_rows3, num_columns3 = df3.shape

print(f"Number of rows: {num_rows1, num_rows2, num_rows3}")
print(f"Number of columns: {num_columns1, num_columns2, num_columns3}")
```

### 2. NaN값을 NULL로 처리

```python
import pandas as pd
import numpy as np

# NaN 값을 'NULL' 문자열로 대체
df1.fillna('Null', inplace=True)
df2.fillna('Null', inplace=True)
df3.fillna('Null', inplace=True)
```

```python
df1
```

```python
df2
```

```python
df3
```

### 3. 일단 concat : 교양(df1) -> 전공(df3) -> 연계창의융합(df2)

```python
# 헤더 제거후 concat
df2.columns = range(df2.shape[1])
df3.columns = range(df3.shape[1])
```

```python
df2
```

```python
df3
```

```python
num_rows1, num_columns1 = df1.shape
num_rows2, num_columns2 = df2.shape
num_rows3, num_columns3 = df3.shape

print(f"Number of rows: {num_rows1, num_rows2, num_rows3}")
print(f"Number of columns: {num_columns1, num_columns2, num_columns3}")
```

```python
df_result = pd.concat([df1, df3, df2], ignore_index=True)

# 결합된 데이터프레임 출력
print(df_result)
```

```python
num_rows, num_columns = df_result.shape
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_columns}")
```

```python
df_result.to_csv(r"C:\Users\user\Desktop\df_result.csv", encoding='cp949')

print("저장 완료")
```

```python
from db import db_connect

con = db_connect()

q = """CREATE VIEW LectureDetailView AS
SELECT
    l.lectureID,
    l.year,
    l.semester,
    l.lecNumber,
    l.lecName,
    l.lecProfessor,
    l.lecClassification,
    l.lecTheme,
    l.lecCredit,
    l.lecTime,
    l.lecWeekTime,
    l.lecClassroom,
    l.isLecClose,
    ld.takenPeople1yearsAgo,
    ld.takenPeople2yearsAgo,
    ld.takenPeople3yearsAgo,
    ld.ForeignLanguage,
    ld.percentageOfOnline,
    ld.isPNP,
    ld.isEngineering,
    ld.isTBL,
    ld.isPBL,
    ld.isSeminar,
    ld.isSmall,
    ld.isConvergence,
    ld.isTeamTeaching,
    ld.isFocus,
    ld.isExperimentDesign,
    ld.isELearning,
    ld.isArt,
    ld.representCompetency,
    ld.learningGoalNmethod,
    ld.Overview,
    ld.evaluationRatio,
    ld.mainBook,
    ld.scheduleNcontent,
    le.everytimeURL,
    le.star,
    le.assignmentAmount,
    le.teamPlayAmount,
    le.gradeAmount,
    le.reviewSummary,
    le.checkAttend,
    le.testNum
FROM
    LectureList l
LEFT JOIN
    LectureDetailData ld ON l.lectureID = ld.lectureID
LEFT JOIN
    LectureEverytimeData le ON l.lectureID = le.lectureID;"""



con.execute(q)
con.commit()  # 변경사항을 저장합니다.
con.close()   # 연결을 닫습니다.
```

```python

def check_lecture_detail_view():
    con = db_connect()
    cursor = con.cursor()

    # 뷰 구조 확인
    cursor.execute("PRAGMA table_info(LectureDetailView)")
    columns = cursor.fetchall()
    print("LectureDetailView 구조:")
    for column in columns:
        print(f"- {column[1]} ({column[2]})")

    # 뷰에서 데이터 조회 (처음 5개 행)
    cursor.execute("SELECT * FROM LectureDetailView LIMIT 5")
    rows = cursor.fetchall()
    print("\nLectureDetailView 샘플 데이터 (처음 5개 행):")
    for row in rows:
        print(row)

    # 특정 조건으로 데이터 조회 (예: 온라인 비율이 50% 이상인 강의)
    cursor.execute("""
    SELECT lectureID, lecName, lecProfessor, percentageOfOnline
    FROM LectureDetailView
    WHERE percentageOfOnline >= 50
    ORDER BY percentageOfOnline DESC
    LIMIT 5
    """)
    online_lectures = cursor.fetchall()
    print("\n온라인 비율이 50% 이상인 강의 (상위 5개):")
    for lecture in online_lectures:
        print(lecture)

    con.close()


check_lecture_detail_view()
```

```python
import sqlite3

DATABASE = './kwu-lecture-database-v8.db'


def db_connect():
    conn = sqlite3.connect(DATABASE)
    conn.row_factory = sqlite3.Row
    return conn


def get_lecture_table_schema():
    conn = db_connect()
    cursor = conn.cursor()

    # LectureTable의 스키마 정보를 가져오는 쿼리
    cursor.execute("PRAGMA table_info(LectureTable)")

    # 결과 출력
    print("LectureTable 스키마:")
    print("--------------------")
    for column in cursor.fetchall():
        print(f"칼럼명: {column['name']}, 데이터 타입: {column['type']}")

    conn.close()


if __name__ == "__main__":
    get_lecture_table_schema()
```
